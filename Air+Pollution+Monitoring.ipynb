```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime 
import matplotlib.dates as md
```

```python
df=pd.read_csv('Desktop/pollution.csv')
df.head()
```

```python
df.drop("entry_id",axis=1,inplace=True)
df.head()
```

```python
 #finding out the time when max. pollution occurs at Chadni Chowk
df[df['Ajmeri Gate']==3.0235897439999997].drop(['Badi chaupar','JLN'],axis=1)
```

```python
 #finding out the time when max. pollution occurs at Gariahat
df[df['Badi chaupar']==3.0235897439999997].drop(['Ajmeri Gate','JLN'],axis=1)
```

```python
 #finding out the time when max. pollution occurs at Dharmatala
df[df['JLN']==3.0235897439999997].drop(['Ajmeri Gate','Badi chaupar'],axis=1)
```

```python
#finding out pollution level at Chadni Chowk by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['created_at'])
y=df['Ajmeri Gate']
plt.plot_date(x, y)
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at Ajmeri Gate")
```

```python
#finding out pollution level at Gariahat by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['created_at'])
y=df['Badi chaupar']
plt.plot_date(x, y,color='green')
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at Badi chaupar")
```

```python
#finding out pollution level at Dharmatala by hourly update
plt.figure(figsize=(12,8))
plt.rcParams.update({'font.size':15})
x =pd.to_datetime(df['created_at'])
y=df['JLN']
plt.plot_date(x, y,color='red')
plt.xlabel("Data & Hour")
plt.ylabel("Pollution Level")
plt.title("Hourly Pollution Level at jln")
```
